```python
import pandas as pd

import numpy as np

import scipy as sp


from scipy import stats

from scipy.stats import norm

from scipy.stats import poisson

import matplotlib.pyplot as plt

```

```python
np.random.seed(312)
data = np.random.randn(200)
plt.hist(data)

plt.xlabel("some data")
plt.ylabel("count");
```

```python
mean, stdev = sp.stats.norm.fit(data)

mean, stdev
```

```python
np.random.seed(312)
data= np.random.randn(200)

plt.hist(data, density= True)
xr = np.linspace(-3, +3, 100)
xr

plt.plot(xr, norm.pdf(xr, loc= 0 , scale= 1))

plt.xlabel("some_data")
plt.ylabel("frequesncy");
```

```python
l = 1

data = np.random.poisson(l, 200)


xr = np.linspace(0, 5, 1)
xr

plt.hist(data)


plt.plot(xr, poisson.pmf(xr, mu=l))


plt.xlabel("some_data")
plt.ylabel("frequesncy");
```

```python
l = 1

data = np.random.poisson(l, 200)


xr = np.arange(0, 5, 1)
xr

plt.hist(data, density= 1)


plt.plot(xr, poisson.pmf(xr, mu=l))


plt.xlabel("some_data")
plt.ylabel("frequesncy");
```

```python
l = 1

data = np.random.poisson(l, 200)


xr = np.arange(0, 5, 1)
xr

plt.hist(data, density= 1)

mean, stdev = sp.stats.norm.fit(data)

plt.plot(xr, norm.pdf(xr, loc=mean, scale= stdev))


plt.plot(xr, poisson.pmf(xr, mu=l))


plt.xlabel("some_data")
plt.ylabel("frequesncy");
```

```python
l = 10

data = np.random.poisson(l, 200)


xr = np.arange (data.min(), data.max(), 1).astype(int)

xr_continous = np.linspace(data.min(), data.max(),100)

plt.hist(data, density= 1)

mean,stdev= sp.stats.norm.fit(data)




plt.plot (xr_continous, norm.pdf(xr_continous, loc= mean, scale= stdev), label = "norm")




plt.plot(xr, poisson.pmf(xr, mu=l), label= "true")


plt.xlabel("some_data")
plt.ylabel("count");
```
