**IDEA**:

Subscribers are expected to have a significantly higher average trip duration in minutes compared to Customers during month May of 2019.

**Data**:

The Citibike data includes information on each customer's trip, including the date, duration (in minutes), and user type. Users are classified into two types: Subscriber and Customer. The dataset for this analysis is limited to trips taken in May 2019.



**Null Hypothesis (H0)**:


For the month of May 2019, the average trip duration in minutes for Subscribers is the same as or less than the average trip duration for Customers.

# _



# _$H_0$_ : $\frac{S_{\mathrm{tripduration}}}{S_{\mathrm{total}}} <= \frac{C_{\ {tripduration}}}{C_{\mathrm{total}}}$



**Alternative Hypothesis (H1):**


For the month of May 2019, the average trip duration in minutes for Subscribers is significantly higher than the average trip duration for Customers.



# _$H_1$_ : $\frac{S_{\mathrm{tripduration}}}{S_{\mathrm{total}}} > \frac{C_{\ {tripduration}}}{C_{\mathrm{total}}}$



**Variables:**


User Type (usertype): Categorical (Subscriber, Customer)
Trip Duration (tripduration): Continuous (in minutes)
Date: Date/Time (focus on May 2019)


**Statistical Test:**


An Independent Samples t-Test will be used to compare the mean trip durations between Subscribers and Customers for May 2019. This test will help determine if there is a significant difference in average trip duration between the two user types within the specified month.
